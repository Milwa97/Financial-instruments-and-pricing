```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## MA(1)

$$ x_t = \epsilon_t + \beta \epsilon_{t-1}$$

where
$\epsilon - WN(0, \sigma^2)$

```python
n = 100
mu = 0
sigma = 1
epsilon = np.random.normal(mu, sigma, n)
```

```python
beta = 0.5
x = [0]
for i in range(1,n):
    x.append( epsilon[i] + beta* epsilon[i-1] )

fig, axs = plt.subplots(2, 1, figsize = (15, 8), sharex = True)

for ax in axs:
    ax.grid(True, alpha = 0.5)
    
axs[0].plot(x, color = 'green')
axs[0].set_ylabel('x(t)', fontsize = 14)

axs[1].plot(epsilon, color = 'blue')
axs[1].set_ylabel('epsilon(t)', fontsize = 14)
axs[1].set_xlabel('time', fontsize = 14)


fig.suptitle('beta = {:}'.format(beta), fontsize = 18)
```

```python
beta = -0.5
x = [0]
for i in range(1,n):
    x.append( epsilon[i] + beta* epsilon[i-1] )

fig, axs = plt.subplots(2, 1, figsize = (15, 8), sharex = True)

for ax in axs:
    ax.grid(True, alpha = 0.5)
    
axs[0].plot(x, color = 'green')
axs[0].set_ylabel('x(t)', fontsize = 14)

axs[1].plot(epsilon, color = 'blue')
axs[1].set_ylabel('epsilon(t)', fontsize = 14)
axs[1].set_xlabel('time', fontsize = 14)


fig.suptitle('beta = {:}'.format(beta), fontsize = 18)
```
